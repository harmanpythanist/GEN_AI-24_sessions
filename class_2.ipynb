```python
import numpy as np

# study hours and exam scores
sh = np.array([2, 3, 4.5, 5, 6.75, 7, 8])  
marks = np.array([60, 67, 70, 75, 40, 85, 90]) 

# Calculate correlation coefficient
correlation_matrix = np.corrcoef(sh, marks)

correlation_matrix
```

```python
correlation = correlation_matrix[0][1]  # study hour and marks relation

print("Correlation Coefficient between study hours and exam scores:", correlation)
```
